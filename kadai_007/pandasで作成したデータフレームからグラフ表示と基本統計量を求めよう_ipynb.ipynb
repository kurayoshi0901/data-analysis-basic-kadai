```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

df = pd.read_csv("sample_pandas_6.csv")
category_df =pd.read_csv("category.csv")
category_df
#データーのマージ
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

#カテゴリー列の要素(弁当、デザートなど)の出現頻度の算出
df_ = df['カテゴリー'].value_counts()

# 棒グラフの作成
plt.figure(figsize=(10, 6))
df_.plot(kind='bar', color='skyblue', edgecolor='black')

# グラフのタイトルとラベルの設定
plt.title('カテゴリー別の頻度', fontsize=16)
plt.xlabel('カテゴリー', fontsize=16)
plt.ylabel('頻度', fontsize=16)

# グラフの表示
plt.tight_layout()
plt.show()

#商品番号列の商品番号ごとに、注文数の基本統計量
df.groupby('商品番号')['注文数'].describe()
```
