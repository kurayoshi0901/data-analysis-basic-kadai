```python

import pandas as pd
import numpy as np

# 問題1
print("問題1")
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
print("df1:")
display(df1)
print("\ndf2:")
display(df2)
print("\n以上の2つのDataFrameを縦方向に結合してください。")

# 解答1
result1 = pd.concat([df1, df2], axis=0)
print("縦方向に結合した結果:")
display(result1)
```

```python
# 問題2
print("問題2")
df3 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df4 = pd.DataFrame({'id': [1, 2, 4], 'age': [25, 30, 35]})
print("df3:")
display(df3)
print("\ndf4:")
display(df4)
print("\n以上の2つのDataFrameを'id'列をキーにしてマージしてください。")

# 問題2 解答
print("'id'列をキーにしてマージした結果:")
result2 = pd.merge(df3, df4, on='id', how='inner')
display(result2)
```

```python
# 問題3
print("問題3")
df5 = pd.DataFrame({
    'category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'value': [10, 20, 30, 40, 50, 60]
})
print("df5:")
display(df5)
print("\n'category'ごとの'value'の平均を計算してください。")

# 問題3 解答
result3 = df5.groupby('category')['value'].mean()

print("'category'ごとの'value'の平均:")
display(result3)
```

```python
# 問題4
print("問題4")
df6 = pd.DataFrame({
    'date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'product': ['A', 'B', 'A', 'B'],
    'sales': [100, 150, 200, 250]
})
print("df6:")
display(df6)
print("\n'date'を行、'product'を列、'sales'を値とするピボットテーブルを作成してください。")

#問題4 解答

pivot_table = df6.pivot(index='date', columns='product', values='sales')

print("ピボットテーブル:")
display(pivot_table)
```
