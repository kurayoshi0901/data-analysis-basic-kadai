```python
#
#あなたは学校の成績管理システムを作成することになりました。以下の要件を満たすPythonプログラムを作成してください。

'''1.以下の学生と成績（3つのテスト結果）を持つ辞書 grades を作成してください：

田中: [80, 75, 90]

佐藤: [65, 70, 85]

鈴木: [90, 95, 88]

高橋: [72, 68, 81]'''

# 学生の成績（辞書形式）
grades = {
    "Alice": [85, 90, 92],
    "Bob": [78, 82, 88],
    "Charlie": [95, 97, 93],
    "David": [67, 75, 80],
    "Eve": [88, 91, 85]
}

# 成績を表示
for student, scores in grades.items():
    print(f"{student}: {scores}")
```

```python
#2.成績の平均点を計算する関数 calculate_average(scores) を作成してください。
def calculate_average(scores):
    if len(scores) == 0:
        return 0
    return sum(scores) / len(scores)

```

```python
#3.合格した学生（平均点が特定の点数以上）をリストアップする関数 get_passed_students(grades, passing_score) を作成してください。
def get_passed_students(grades, passing_score):
    passed_students = []
    for student, scores in grades.items():
        average_score = calculate_average(scores)
        if average_score >= passing_score:
            passed_students.append(student)
    return passed_students

```

```python
# 4.新しい学生と成績を追加する関数 add_new_student(grades, student, scores) を作成してください。
def add_new_student(grades, student, scores):
    grades[student] = scores
```

```python
# 5.全生徒の成績と平均点を表示する関数 print_all_grades(grades) を作成してください
def print_all_grades(grades):
    for student, scores in grades.items():
        average_score = calculate_average(scores)
        print(f"{student}: Scores = {scores}, Average = {average_score:.1f}")#小数点第１位まで表示
```

```python
#メイン処理として、以下の操作を行ってください：

''''a. 現在の全生徒の成績と平均点を表示する

b. 合格点を75点とし、合格した学生をリストアップして表示する

c. 新しい学生 "山田" と成績 [78, 82, 88] を追加する

d. 更新後の全生徒の成績と平均点を表示する'''

# a. 現在の全生徒の成績と平均点を表示する
print("現在の全生徒の成績と平均点:")
print_all_grades(grades)
print("\n")

# b. 合格点を75点とし、合格した学生をリストアップして表示する
passing_score = 75
passed_students = get_passed_students(grades, passing_score)
print(f"合格者(passing score: {passing_score}):")
print(passed_students)
print("\n")

# c. 新しい学生 "山田" と成績 [78, 82, 88] を追加する
add_new_student(grades, "山田", [78, 82, 88])

# d. 更新後の全生徒の成績と平均点を表示する
print("更新後の全生徒の成績と平均点:")
print_all_grades(grades)
```

```python
#1. 複数のプロット・マーカ等の変更
'''データ作成。以下のコードでデータを作成してください。

import numpy as np

x = np.linspace(0, 10, 100)

y1 = np.sin(x)

y2 = np.cos(x)

y3 = np.tan(x)
以下を実現してください
折れ線グラフで3種類の線を1つのグラフに描画してください。

リスト項目

リスト項目

マーカー・線の色・線のスタイル・ラベルをそれぞれ変えてプロットしてください。

凡例を表示してください。

画像サイズ(figsize)をデフォルトから変更してください。'''

import numpy as np
import matplotlib.pyplot as plt

# データ作成
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# グラフサイズを設定
plt.figure(figsize=(10, 6))

# y1のプロット (sin)
plt.plot(x, y1, marker='o', color='blue', linestyle='-', label='sin(x)')

# y2のプロット (cos)
plt.plot(x, y2, marker='x', color='green', linestyle='--', label='cos(x)')

# y3のプロット (tan)
plt.plot(x, y3, marker='s', color='red', linestyle=':', label='tan(x)', alpha=0.7)

# グラフのタイトルとラベル
plt.title('Trigonometric Functions', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# 凡例を表示
plt.legend()

# グリッドを表示
plt.grid(True)

# グラフの表示
plt.show()
```
